```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#Sample Data
data={'Country': ['China', 'India', 'United States', 'Indonesia', 'Pakistan',
                'Brazil', 'Nigeria', 'Bangladesh', 'Russia', 'Mexico'],
    'Population': [1439, 1380, 331, 273, 225,213, 211, 166, 146, 129],
    'Continent': ['Asia', 'Asia', 'North America', 'Asia', 'Asia',
                  'South America', 'Africa', 'Asia', 'Europe', 'North America']
}
#Color map for continents
df=pd.DataFrame(data)
continent_colors={'Asia':'Red','Africa':'Green','Europe':'Blue','North America':'Yellow','South America':'Purple'}
#Create bar plt
plt.figure(figsize=(12,6))
sns.barplot(data=df,x='Country',y='Population',hue='Continent',palette='viridis',dodge=False)
plt.title('Country by population(in millions)')
plt.xlabel('Country')
plt.ylabel('Population(in millions)')
plt.xticks(rotation=45)
plt.legend(title='Continent')
plt.tight_layout()
plt.show()
plt.figure(figsize=(10,6))
bin=[0,50,200,500,1000,1500]
sns.histplot(data=df,x='Population',bins=bin,kde=True,color='Teal',edgecolor='Black')
plt.title('Distribution of country population(with density curve)',fontsize=16)
plt.xlabel('Population( in millions)',fontsize=12)
plt.ylabel('Number of Countries',fontsize=12)
plt.xticks(bin)
plt.grid(axis='y',alpha=0.3)
plt.tight_layout()
plt.show()
years= list(range(2000, 2021))
countries = {
    'China': [1263, 1276, 1288, 1300, 1312, 1324, 1336, 1348, 1360, 1371,
              1383, 1394, 1404, 1414, 1425, 1434, 1443, 1451, 1459, 1467, 1475],
    'India': [1053, 1071, 1089, 1107, 1125, 1143, 1161, 1179, 1197, 1214,
              1232, 1249, 1266, 1283, 1300, 1317, 1334, 1350, 1366, 1382, 1398],
    'USA': [282, 285, 288, 291, 294, 297, 300, 303, 306, 309,
            311, 314, 316, 318, 320, 322, 324, 326, 328, 330, 332]}
plt.figure(figsize=(12,6))
for country,pop_data in countries.items():
  plt.plot(years,pop_data,marker='o',label='country',linewidth=0.2)
plt.title('Population growth over time(2000-2020)',fontsize=16)
plt.xlabel('Year',fontsize=12)
plt.ylabel('Population(in millions)',fontsize=12)
plt.grid(True,alpha=0.3)
plt.legend()
#Add data label for first and last year
for country,pop_data in countries.items():
   plt.text(years[0], pop_data[0], f'{pop_data[0]}M', ha='right', va='center')
   plt.text(years[-1], pop_data[-1], f'{pop_data[-1]}M', ha='left', va='center')

plt.show()

data={'Country': ['Monaco', 'Singapore', 'Bahrain', 'Maldives', 'Malta',
                'Bangladesh', 'Palestine', 'Lebanon', 'Barbados', 'Taiwan'],
    'Population': [0.04, 5.7, 1.5, 0.5, 0.5, 165, 5.0, 6.8, 0.3, 23.6],
    'Area': [0.002, 0.7, 0.8, 0.3, 0.3,
             148, 6.0, 10.5, 0.4, 36.2]}
df=pd.DataFrame(data)
df['Density']=df['Population']/df['Area']     #Calculate density
df=df.sort_values('Density',ascending=False)
plt.figure(figsize=(12,6))
heatmap=sns.heatmap(df[['Density']].T,annot=True,fmt='.1f',cmap='YlOrRd',yticklabels=['Population Density'],xticklabels=df['Country'])
plt.title('Population Density by country (people per sqkm)',fontsize=16)
plt.xlabel('Country',fontsize=12)
plt.ylabel('')
plt.xticks(rotation=45)
cbar=heatmap.collections[0].colorbar
cbar.set_label('Population per sqkm',rotation=270,labelpad=20)
plt.tight_layout()
plt.show()
```
